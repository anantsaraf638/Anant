```python
# Install the 'huggingface_hub' package with optional 'hf_xet' support (used for optimized file storage on Hugging Face Hub)
!pip install huggingface_hub[hf_xet]
# Import the summarization pipeline from the Hugging Face Transformers library
from transformers import pipeline

# Define a function to summarize a given article
def summarize_article(text, max_length=150, min_length=40):
    # Load the BART model pre-trained for summarization
    summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
     # Generate the summary using the specified max and min lengths
    summary = summarizer(text, max_length=max_length, min_length=min_length, do_sample=False)
    # Return only the summary text from the result
    return summary[0]['summary_text']
# Input article text to be summarized
article = """
Artificial intelligence (AI) has significantly transformed various sectors including healthcare, finance, transportation, and education.
In healthcare, AI-powered diagnostics help detect diseases earlier and with greater accuracy. In finance, fraud detection algorithms help 
identify suspicious activities in real-time. The transportation industry benefits from AI through autonomous vehicles and traffic prediction systems.
Education has seen a rise in adaptive learning platforms, personalizing education for students based on their learning styles.
Despite these benefits, AI also presents challenges such as data privacy concerns, job displacement due to automation,
and ethical dilemmas around decision-making by machines. Governments and organizations around the world are working to 
establish regulations to ensure responsible AI development and deployment. The future of AI holds immense promise if its
risks are managed thoughtfully.
"""

# Input article text to be summarized
summary = summarize_article(article)

# Print the generated summary

print("=== SUMMARY ===\n")
print(summary)
```
